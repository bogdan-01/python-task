**2. Deal with missing data**

```python
forest_fires = forest_fires.fillna({'area':0})
```

```python
forest_fires.head()
```

```python
forest_fires.dropna(inplace=True)
```

```python
forest_fires.head()
```

**3. Using the "Verbs"**

```python
forest_fires = forest_fires[["month", "day", "temp", "wind", "rain", "area"]]
forest_fires.head() #selected all but X and Y columns, reassigned
```

```python
forest_fires.sort_values("area", ascending = False) #arranged by area, highest value first
```

```python
#filter rows where area is not zero
forest_fires_no_zero = forest_fires[forest_fires["area"] != 0]
```

```python
forest_fires_no_zero
```

```python
# Method 1: I'll create new column 'is_rain' showing 'True' if rain > 0 or 'False' otherwise.
# forest_fires_no_zero['is_rain'] = np.where(forest_fires_no_zero['rain'] > 0, 'True', 'False')
```

```python
# Method 2: Same as above but less code.
forest_fires_no_zero['is_rain'] = forest_fires_no_zero['rain'] > 0
```

```python
forest_fires_no_zero
```

```python
# I'll find the average area of fire in each month
forest_fires_no_zero.groupby('month')['area'].mean()
```

```python
# I'll find the average area of fire in each day
forest_fires_no_zero.groupby('day')['area'].mean()
```
